```r
library(ggplot2)

# Create a sequence of x values
x <- seq(-2*pi, 2*pi, length.out = 100)

# Create a data frame
df <- data.frame(x = x, sin_x = sin(x), cos_x = cos(x))

# Plot sine and cosine functions
ggplot(df, aes(x)) +
  geom_line(aes(y = sin_x, color = "sin(x)")) +
  geom_line(aes(y = cos_x, color = "cos(x)")) +
  labs(title = "Sine and Cosine Functions", x = "x", y = "y") +
  scale_color_manual(values = c("sin(x)" = "blue", "cos(x)" = "red"))
```

```r
df_step <- data.frame(x = seq(-5, 5, 1), y = ifelse(seq(-5, 5, 1) > 0, 1, 0))

ggplot(df_step, aes(x, y)) +
  geom_step() +
  labs(title = "Step Function", x = "x", y = "y")
```

```r
df_exp <- data.frame(x = seq(-2, 2, length.out = 100), y = exp(seq(-2, 2, length.out = 100)))

ggplot(df_exp, aes(x, y)) +
  geom_line(color = "green") +
  labs(title = "Exponential Function", x = "x", y = "e^x")
```

```r
x <- seq(-10, 10, length.out = 1000)

# Define each function
y_poly <- x^3 - 4*x             # Polynomial
y_abs <- abs(x)                 # Absolute value
x_log <- seq(0.1, 10, length.out = 1000)  # For log and sqrt, avoid x <= 0
y_log <- log(x_log)             # Natural logarithm
y_sqrt <- sqrt(x_log)           # Square root
y_gauss <- exp(-x^2)            # Gaussian function

# Plot 1: Polynomial
plot(x, y_poly, type = "l", col = "blue", main = "Polynomial Function: x^3 - 4x", xlab = "x", ylab = "f(x)")

# Plot 2: Absolute Value
plot(x, y_abs, type = "l", col = "red", main = "Absolute Value Function: |x|", xlab = "x", ylab = "f(x)")

# Plot 3: Logarithmic
plot(x_log, y_log, type = "l", col = "darkgreen", main = "Logarithmic Function: log(x)", xlab = "x", ylab = "f(x)")

# Plot 4: Square Root
plot(x_log, y_sqrt, type = "l", col = "orange", main = "Square Root Function: sqrt(x)", xlab = "x", ylab = "f(x)")

# Plot 5: Gaussian
plot(x, y_gauss, type = "l", col = "purple", main = "Gaussian Function: e^(-x^2)", xlab = "x", ylab = "f(x)")
```

```r
# Create a range of x values
x_arc <- seq(-1, 1, length.out = 500)   # For arcsin and arccos, domain is [-1, 1]
x_atan <- seq(-10, 10, length.out = 500)  # For arctan, domain is all real numbers

# Calculate y-values
y_asin <- asin(x_arc)    # arcsin(x)
y_acos <- acos(x_arc)    # arccos(x)
y_atan <- atan(x_atan)   # arctan(x)

# Plot arcsin
plot(x_arc, y_asin, type = "l", col = "blue", lwd = 2,
     main = "Inverse Sine Function: inversesin(x)", xlab = "x", ylab = "arcsin(x)")

# Plot arccos
plot(x_arc, y_acos, type = "l", col = "red", lwd = 2,
     main = "Inverse Cosine Function: inversecos(x)", xlab = "x", ylab = "arccos(x)")

# Plot arctan
plot(x_atan, y_atan, type = "l", col = "green", lwd = 2,
     main = "Inverse Tangent Function: inversetan(x)", xlab = "x", ylab = "arctan(x)")
```

```r
# Range for x
x_sec_csc <- seq(-5, 5, length.out = 500)
x_sec_csc <- x_sec_csc[x_sec_csc <= -1 | x_sec_csc >= 1]  # secant and cosecant are defined only for |x| >= 1
x_cot <- seq(-10, 10, length.out = 500)

# Define the functions
arcsec <- function(x) { acos(1/x) }
arccsc <- function(x) { asin(1/x) }
arccot <- function(x) { atan(1/x) }

# Calculate y-values
y_arcsec <- arcsec(x_sec_csc)
y_arccsc <- arccsc(x_sec_csc)
y_arccot <- arccot(x_cot)

# Plot arcsec
plot(x_sec_csc, y_arcsec, type = "l", col = "blue", lwd = 2,
     main = "Inverse Secant Function: arcsec(x)", xlab = "x", ylab = "arcsec(x)")

# Plot arccsc
plot(x_sec_csc, y_arccsc, type = "l", col = "red", lwd = 2,
     main = "Inverse Cosecant Function: arccsc(x)", xlab = "x", ylab = "arccsc(x)")

# Plot arccot
plot(x_cot, y_arccot, type = "l", col = "green", lwd = 2,
     main = "Inverse Cotangent Function: arccot(x)", xlab = "x", ylab = "arccot(x)")
```

```r
# Set up plotting area: 1 row, 2 columns
par(mfrow = c(1, 2))

# Define x values
x_mod <- seq(-10, 10, length.out = 500)
x_inv <- seq(-10, 10, length.out = 500)
x_inv <- x_inv[x_inv != 0]  # exclude 0 to avoid division by zero

# 1. Modulus Function: |x|
plot(x_mod, abs(x_mod), type = "l", col = "blue", lwd = 2,
     main = "Modulus Function: |x|", xlab = "x", ylab = "f(x)")

# 2. Reciprocal Function: 1/x
plot(x_inv, 1/x_inv, type = "l", col = "red", lwd = 2,
     main = "Reciprocal Function: 1/x", xlab = "x", ylab = "f(x)")
abline(h = 0, v = 0, lty = 2, col = "gray")  # Add asymptotes
```

```r
# 1. All functions from A to B: n^m
num_all_functions <- function(m, n) {
  return(n^m)
}

# 2. One-one functions: n! / (n - m)!
num_one_one_functions <- function(m, n) {
  if (m > n) {
    return(0)
  }
  return(factorial(n) / factorial(n - m))
}

# 3. Onto functions using inclusion-exclusion
num_onto_functions <- function(m, n) {
  if (n > m) {
    return(0)
  }
  total <- 0
  for (k in 0:n) {
    total <- total + (-1)^k * choose(n, k) * (n - k)^m
  }
  return(total)
}
```
